## Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df_train = pd.read_csv("./train.csv")

df_train.columns = df_train.columns.str.replace(' ','_')
```

```python
df_train.head(15)
```

```python
df_train.info()
```

## Time Spent Alone

```python
df_train['Time_spent_Alone'] = df_train['Time_spent_Alone'].astype('Int64')
df_train['Time_spent_Alone'] = df_train['Time_spent_Alone'].fillna(1000)
tsa = df_train.groupby(['Time_spent_Alone', 'Personality']).size().reset_index(name='Count')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=tsa, x='Time_spent_Alone', y='Count', hue='Personality')

plt.title('Personality type by Time spent Alone')
plt.xlabel('Time spent alone')
plt.ylabel('Number of people')
plt.tight_layout()
plt.show()
```

Wnioski:
- zmieni floaty na inty
- fillna - uzupelnij 2 (bo wiekszosc NaN to extraverts)
